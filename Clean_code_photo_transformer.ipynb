```python
from PIL import Image
import numpy as np
import pandas as pd
import glob
```

```python
df=pd.read_csv("Train.csv")
df = df.drop('Unnamed: 0', 1)
df
```

```python
img =[Image.open(file).convert('L') for file in glob.glob(r"C:\Users\Huawei\Desktop\boots\*.jpg")]
```

```python
for image in img:
    resized_img = image.resize((28, 28))
    data_3d = np.asarray(resized_img, dtype="int32")
    data = data_3d.reshape(-1)
    new_series =pd.Series(data, index = df.columns)
    #transed.append(new_series)
    df = df.append(new_series, ignore_index=True)
```
